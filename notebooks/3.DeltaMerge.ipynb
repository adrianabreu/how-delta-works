```scala
val sampleData = Seq(
    (1,"Pedro"),
    (2,"Sergi"),
    (3,"Aleix"),
    (4,"David")
)

import spark.implicits._
val df = sampleData.toDF()

df.coalesce(1).write.mode("overwrite").format("delta").saveAsTable("data_engineers_merge")

import io.delta.tables._

val newDataDf = Seq(
    (4,"Adrian"),
    (5,"David"),
    (6,"Simon")
).toDF()

DeltaTable
.forName("data_engineers_merge")
.as("oldData")
  .merge(
    newDataDf.as("newData"),
    "oldData._1 = newData._1")
  .whenMatched
  .updateAll
  .whenNotMatched
  .insertAll
  .execute()
```
